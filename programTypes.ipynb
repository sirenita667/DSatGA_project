```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

##plotting program types (2-letter designation) against time and funding amounts

```python
d = pd.read_csv('https://raw.githubusercontent.com/sirenita667/DSatGA_project/master/imls.csv')
```

```python
#type(float('$750000.00'.replace('$', "")))

aa = d['Total Amount Awarded and Disbursed']

for a in aa:
    aa = aa.replace(a, float(a.replace('$', "")))
print aa[0:4]
```

```python


from datetime import datetime

#09/26/2014 12:00:00 AM



bb = []

for b in d['Award Date']:
    try:
        if np.isnan(b):
            bb.append(b)
        else:
            bb.append(datetime.strptime(b, '%m/%d/%Y %I:%M:%S %p'))
    except:
        bb.append(datetime.strptime(b, '%m/%d/%Y %I:%M:%S %p'))
        
print len(aa)
print bb[0:4]
print len(bb)
print len(d)
```

```python
d['totAward_float'] = pd.Series(aa)
d['date_asTime'] = pd.Series(bb)
```

```python
list(d)
```

```python
#taking out records were d.date_asTime are NaN

d1 = d[~pd.isnull(d.date_asTime)]

print(len(d), len(d1))
```

```python
#IA -- Conservation Assessment Program

pt = d1['Program Type']


xIA = np.array(d1[(pt =='IA')].date_asTime)
yIA = np.array(d1[(pt =='IA')].totAward_float)

plt.show()
plt.scatter(xIA, yIA, c='pink')
plt.xlabel('year')
plt.ylabel('$')
plt.title('Conservation Assessment Program (IA) -- Full')

plt.show()
plt.scatter(xIA, yIA, c = 'pink')
plt.ylim(ymax=8500, ymin= 2500)
plt.xlim(xmin=datetime.strptime('1995', '%Y'), xmax = datetime.strptime('2007', '%Y'))
plt.xlabel('year')
plt.ylabel('$')
plt.title('Conservation Assessment Program (IA) -- Zoomed')

```
